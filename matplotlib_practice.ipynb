```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis,y_axis, label = "Boston", marker = "d", color = "green", linewidth = 2)

plt.xlabel("Date",weight="bold")
plt.ylabel("Fare($)",weight="bold")

plt.ylim(0,45)

plt.title("PyBer Fare by Month",weight="bold")

plt.grid()

plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(x_axis,y_axis, color ="green", label = "Boston")

plt.xlabel("Month")
plt.ylabel("Fare($)")

plt.title("PyBer Fare by Month")

plt.legend()
```

```python
plt.barh(x_axis,y_axis, label = "Boston", color = "magenta")
plt.gca().invert_yaxis()

plt.legend()

plt.title("Pyber Fare by Month")

plt.xlabel("Fare ($)")
plt.ylabel("Month")

```

```python
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
plt.barh(x_axis, y_axis, label = "Chicago", color = "cyan")

plt.xlabel("Date")
plt.ylabel("Fare ($)")

plt.legend()
```

```python
plt.plot(x_axis,y_axis,"o")
```

```python
plt.scatter(y_axis,x_axis, c="red", label = "Chicago", s = y_axis)

plt.legend()

plt.title("PyBer Fare by Month")

plt.xlabel("Date")
plt.ylabel("Fare ($)")

plt.gca().invert_yaxis()

plt.xlim(0,50)
```

```python
plt.scatter(x_axis,y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i * 5 for i in y_axis], c="skyblue", edgecolor = "black",
           linewidth = 2, alpha = .8, label = "Boston")

plt.title("PyBer Fare by Month")

plt.xlabel("Fare ($)")
plt.ylabel("Month")

plt.xlim(0,50)

plt.legend()

plt.gca().invert_yaxis()

fig.set_size_inches(7,5)
```

```python
plt.pie(y_axis,labels = x_axis)
plt.show()
```

```python
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, labels = x_axis, autopct = "%.1f%%", explode = explode_values, radius = 2, colors = colors, shadow = True, startangle = 90)
plt.gca().invert_xaxis()
plt.show()
```
